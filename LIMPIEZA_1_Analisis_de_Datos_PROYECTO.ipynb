```python
!pip install sidetable
import pandas as pd
import numpy as np
import sidetable as stb

pd.options.display.max_columns = None
```

```python
df = pd.read_csv("/content/COPIA- atenciones_oficina_publica_de_empleo_ope.csv")
df.head(100)
```

```python
# Comprobamos los nulos
df.stb.missing()
```

```python
# Eliminamos la columna zona cuidadano ya que mas del 90% de los datos son nulos

#df["zona_ciudadano"].replace(np.nan, "Unknown", inplace = True)

df = df.drop(columns=["zona_ciudadano"])
```

```python
df.shape
```

```python
df.dtypes
```

```python
# Llenamos los campos vacios de edad con la media

df["edad"].fillna(round(df["edad"].mean(),2), inplace = True)
# Luego convertimos este tipo de dato en entero

df['edad'] = df['edad'].astype(int)
```

```python
df.edad.isnull().sum()
```

```python
#Buscamos si los datos nulos fueron corregidos en la totalidad de las columnas
```

```python
df.stb.missing()
```

```python
# Vamos a convertir las columnas 'fecha_de_beneficio_dia'... mes y año en una sola columna que se llame 'fecha_de_beneficio'

df['fecha_de_beneficio'] = pd.to_datetime(df['fecha_de_beneficio_año'].astype(str) + '-' + df['fecha_de_beneficio_mes'].astype(str) + '-' + df['fecha_de_beneficio_dia'].astype(str))
```

```python
# Revisamos que quede correcta la columna fecha_de_beneficio
df.head(5)
```

```python
# Eliminamos las columnas con el dia, mes y año

df.drop(columns=["fecha_de_beneficio_dia", "fecha_de_beneficio_mes", "fecha_de_beneficio_año"], inplace = True)
```

```python
df.head(5)
```

```python
# Vemos datos duplicados en la columna 'nombre_del_proyecto'
df.nombre_del_proyecto.duplicated().sum()
```

```python
# Eliminamos la columna ya que todos estos datos están repetidos y es innecesario tenerlos

df.drop(columns=["nombre_del_proyecto"], inplace = True)
```

```python
# Modificamos el tipo de dato de la columna 'estrato' a entero
df['estrato'] = pd.to_numeric(df['estrato'], errors='coerce').astype('Int64')
```

```python
df.estrato.isnull().sum()
```

```python
# Llenamos datos vacios con la media

df["estrato"].fillna(round(df["estrato"].mean()), inplace = True)
```

```python
df.dtypes
```

```python
#Descargamos la data limpia con el prefijo CLEAN en el nombramiento
```

```python
df.to_csv("/content/CLEAN-atenciones_oficina_publica_de_empleo_ope.csv")
```
